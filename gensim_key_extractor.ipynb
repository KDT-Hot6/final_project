## Unsupervised topic model and using TextRank algorithm

```
from gensim.summarization import keywords
```

```
import pandas as pd
import re

data = pd.read_csv('/content/drive/MyDrive/Colab_Notebooks/menupan_crawling.csv')
data.head()
```

```
k = 0
## 정규식 : 한글, 숫자만 추출
for i in data['review']:
  text = re.compile('[ㄱ-ㅎ|\d\ㅏ-ㅣ|가-힣]+').findall(i)
  data.loc[k, 'review'] = ' '.join(text).strip()
  k+=1
```

```
## review 특정 단어 제외
data['review'] = data['review'].str.replace(pat=r'[ㅋㅋㅋ, 좋아요]', repl=r' ', regex=True)
data.head()
```

```
def key_extract(text):
  return keywords(text, words= 5).split('\n')
```

```
for i in range(len(data[:200])):
    print(data.loc[i,'res_name'], ':', key_extract(data.loc[i,'review']))
```
