```python
df_d = pd.read_csv('./data/merge_table/review_diningcode.csv')
df_d
```

```python
df_d.drop(['Unnamed: 0','평점','공감수','특징'], axis='columns', inplace=True)
df_d
```

```python
df_d = df_d.rename(columns={'식당이름':'RES_NAME', '리뷰':'REV_COMMENT','주소':'RES_ADDRESS'})
df_d['PRE'] = 'D'
df_d['REV_PRO_COM'] = df_d['REV_COMMENT'].str.replace(pat=r'[^가-힣]', repl=r' ', regex=True)
df_d
```

```python
df_d['RES_ID'] = 1
df_d
```

```python
df_d['UNI'] = df_d['RES_NAME'] + df_d['RES_ADDRESS']
df_d
```

```python
uni = df_d['UNI'].unique()
```

```python
d = {}
cnt = 1
for u in uni:
    d[u] = cnt
    cnt += 1
d
```

```python
df_d['RES_ID'] = df_d['UNI'].apply(lambda x:d[x])
df_d
```

```python
df_d['RES_ID'] = df_d['RES_ID'].astype(str)
df_d['RES_ID_NEW'] = df_d['PRE'] + df_d['RES_ID']
df_d
```

```python
df_p['RES_ID'] = 1
df_p['UNI'] = df_p['RES_NAME'] + df_p['RES_ADDRESS']
uni = df_p['UNI'].unique()
d2 = {}
cnt = 1
for u in uni:
    d2[u] = cnt
    cnt += 1
df_p['RES_ID'] = df_p['UNI'].apply(lambda x:d2[x])
df_p
```

```python
df_p['RES_ID'] = df_p['RES_ID'].astype(str)
df_p['RES_ID_NEW'] = df_p['PRE'] + df_p['RES_ID']
df_p
```
