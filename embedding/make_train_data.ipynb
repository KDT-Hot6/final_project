```python
long = {'sent_1':[],'sent_2':[],'score':[]}
short = {'sent_1':[],'sent_2':[],'score':[]}
```

```python
import pandas as pd
```

```python
df = pd.read_csv('../data/make_train_review.csv')
df.head()
```

```python
df['PRE_COMMENT'] = df['REV_COMMENT'].str.replace(pat=r'[^가-힣]', repl=r' ', regex=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(axis=0,inplace=True)
df.isnull().sum()
```

```python
short2 = []
for i, row in df.iterrows():
    try:
        long_sent = str(row['REV_COMMENT'])
        summ = summarize(long_sent)
        # long에 넣기
        long['sent_1'].append(str(row['PRE_COMMENT']))
        long['sent_2'].append(summ)
        long['score'].append(4.5)
    except:
        # short에 넣기
        short2.append(str(row['PRE_COMMENT']))
```

```python
df_long = pd.DataFrame(long)
df_long.head()
```

```python
df_long['pro_sent_2'] = df_long['sent_2'].str.replace(pat=r'[^가-힣]', repl=r' ', regex=True)
df_long
```
