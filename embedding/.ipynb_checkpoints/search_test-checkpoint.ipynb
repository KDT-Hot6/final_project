```python
stop_words = ['은','는','이','가','한','를','을','곳','집','식당','장소','하','기']
def get_keyword_token(sentence):
    processed_sentence = tagger.morphs(sentence)
    res = []
    for item in processed_sentence:
        if item not in stop_words:
            res.append(item)
    return res
```

```python
print('모임하기 좋은 식당:',get_keyword_token('모임하기 좋은 식당'))
print('분위기 좋은곳',get_keyword_token('분위기 좋은곳'))
print('육즙팡팡 스테이크',get_keyword_token('육즙팡팡 스테이크'))
print('고기냄새 안나는 족발집',get_keyword_token('고기냄새 안나는 족발집'))
print('프라이빗한 식당',get_keyword_token('프라이빗룸'))
```

```python
def get_keyword_vector(keyword):
    similar_list = review_only_model.most_similar(keyword, topn= 300)
    words = [keyword]
    res = []
    for item in similar_list:
        if keyword not in str(item[0]):
            res.append(item)
    while len(words) < 5:
        if not res:
            break
        tmp = res.pop(0)
        words.append(tmp[0])
        tmp_res = []
        for item in res:
            if tmp[0] not in str(item[0]):
                tmp_res.append(item)
        res = tmp_res    
    return words
```

```python
def get_query(tokens,table_name):
    sql = 'SELECT * FROM `'+table_name+'` WHERE {};'
    sub = ''
    for i,token_list in enumerate(tokens):
        if i != 0:
            sub += ' AND '
        sub += '({})'
        form = '(`REV_COMMENT` LIKE \'%{}%\')'
        sub2 = ''
        for j,token in enumerate(token_list):
            if j != 0:
                sub2 += ' OR '
            sub2 += form.format(token)
        sub = sub.format(sub2)
    return sql.format(sub) 
```

```python
tokens = [['프라이빗', '테라스와', '독립된', '라이브', '파티션이'],['룸', '칸막이', '단체석', '다찌석', '예약제']]
print(get_query(tokens))
```

# 검색어 토큰 함수

```python
stop_tags = ['JKS','JKC','JKG','JKO','JKB','JKV','JKQ','JX','JC','EP','EF','EC','ETN','ETM','XSN','XSV','XSA','SF','SE','SSO','SSC','SC','SY']
stop_words = ['한','곳','집','식당','장소','음식점']
def get_search_token(sent):
    spl_tokens = sent.split(' ')
    res = []
    for st in spl_tokens:
        pos_st = tagger.pos(st)
        mor_st = tagger.morphs(st)
        while pos_st and ( pos_st[-1][1] in stop_tags or pos_st[-1][0] in stop_words ):
            pos_st = pos_st[:-1]
            mor_st = mor_st[:-1]
        tmp = ''.join(mor_st)
        if len(tmp) > 1:
            res.append(tmp)
#         if pos_st[-1][1] in stop_tags or pos_st[-1][0] in stop_words:
#             res.append(''.join(mor_st[:-1]))
#         else:
#             res.append(''.join(mor_st))
    return res
```

```python
print('개존맛 존맛탱',get_search_token('개존맛 존맛탱'))
print('겉바속촉 빵',get_search_token('겉바속촉 빵'))
print('매장이 깨끗한곳',get_search_token('매장이 깨끗한곳'))
print('바삭바삭한 감자튀김',get_search_token('바삭바삭한 감자튀김'))
print('분위기쩌는 데이트 맛집',get_search_token('분위기쩌는 데이트 맛집'))
print('고기 구워주는집',get_search_token('고기 구워주는집'))
print('직원이 친절한 닭갈비집',get_search_token('직원이 친절한 닭갈비집'))
print('이영자 맛집',get_search_token('이영자 맛집'))
print('인싸 맛집',get_search_token('인싸 맛집'))
print('회식장소로 괜찮은 곳',get_search_token('회식장소로 괜찮은 곳'))
```

# 벡터 유사도 단어 함수

```python
def get_vector_similar_words(keywords):
    result = []
    for keyword in keywords:
        similar_list = review_only_model.most_similar(keyword, topn= 300)
        words = [keyword]
        res = []
        for item in similar_list:
            if keyword not in str(item[0]):
                res.append(item)
        while len(words) < 5 and res:
            if not res:
                break
            tmp = get_search_token(res.pop(0)[0])
            if tmp:
                tmp = tmp[0]
                words.append(tmp)
                tmp_res = []
                for item in res:
                    if tmp not in str(item[0]) and item[1] > 0.7:
                        tmp_res.append(item)
                res = tmp_res
        result.append(words)
    return result
```
